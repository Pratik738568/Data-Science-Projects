# Performing Preprocessing Step

## 1.Creating over own dataset


```python
import numpy as np
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'country':['india','france','germany','france','germany','india','germany','france','india','france'],\
                   'age':[38,43,30,48,40,35,np.nan,49,50,37],\
                   'salary':[68000,45000,54000,65000,np.nan,58000,53000,79000,88000,77000],\
                   'purchased':['no','yes','no','no','yes','yes','no','yes','no','yes']
})
```

```python
df
```

## 2.Missing Value Imputation

```python
df['age'].mean()
```

```python
df['salary'].mean()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
df_imputed=pd.DataFrame()
```

```python
df_imputed['country']=df['country']
```

```python
df_imputed
```

```python
df_imputed['age']=imp_mean.fit_transform(df[['age']])
```

```python
df['age']
```

```python
df[['age']]
```

```python
df_imputed
```

```python
df_imputed['salary']=imp_mean.fit_transform(df[['salary']])
```

```python
df_imputed['purchased']=df['purchased']
```

```python
df_imputed
```

## 3.Performing Lable Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc=LabelEncoder()
```

```python
df_encoded=pd.DataFrame()
```

```python
df_encoded['country']=enc.fit_transform(df_imputed['country'])
```

```python
df_encoded
```

```python
df_encoded['age']=df_imputed['age']
```

```python
df_encoded['salary']=df_imputed['salary']
```

```python
df_encoded
```

```python
df_encoded['purchased']=enc.fit_transform(df_imputed['purchased'])
```
