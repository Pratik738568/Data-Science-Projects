creating X & y

```python
X=df.drop(['num'],axis=1)
y=df['num']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X_train,y_train)
```

```python
print("training accuracy=",clf.score(X_train,y_train))
```

```python
print("testing accuracy=",clf.score(X_test,y_test))
```

##Our model has high variance i.p.overfitted
Therefore, going for proning

```python
clf2=RandomForestClassifier(max_depth=3)
clf2.fit(X_train,y_train)
print("training accuracy=",clf2.score(X_train,y_train))
print("testing accuracy=",clf2.score(X_test,y_test))
```

```python
clf3=RandomForestClassifier(max_depth=3,class_weight='balanced')
clf3.fit(X_train,y_train)
print("training accuracy=",clf3.score(X_train,y_train))
print("testing accuracy=",clf3.score(X_test,y_test))
```
