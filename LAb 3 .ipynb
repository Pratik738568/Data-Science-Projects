## using labelencoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
```

```python
df_encoded = pd.DataFrame()
```

```python
df_encoded['gender'] = df['gender']
```

```python
df_encoded['race/ethnicity'] = df['race/ethnicity']
```

```python
df_encoded['Class_Type'] = enc.fit_transform(df['Class_Type'])
```

```python
df_encoded['Animal_Names'] = enc.fit_transform(df['Animal_Names'])
```

```python
df_encoded.dtypes
```

```python
df_encoded.columns
```

## Approach 1: Brute Force approach using all features

```python
X=df_encoded.drop(['Class_Number'],axis=1)
```

```python
y=df_encoded[['Animal_Names']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
reg.score (X_test,y_test)
```

```python
reg.coef_
```

```python
reg.intercept_
```
